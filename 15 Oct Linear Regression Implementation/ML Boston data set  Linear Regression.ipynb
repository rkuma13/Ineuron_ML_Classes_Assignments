```python
#Training the point
#Independent features X and Dependent features
# I want all the 13 independent features in X
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
#Train test split
#using sklearn google it

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
   X,y, test_size=0.33, random_state=42)
```

```python
#standardize or feature scaling the datasets

from sklearn.preprocessing import StandardScaler
#in standardscaler mean is 0 and Standard deviation is 1

scaler = StandardScaler()
```

```python
scaler
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

**Model Training**

```python
#This is a multiple linear regression probme, since multiple independent features
#now google sklearn linear model
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression
```

```python
# I have the training data , now train the training data
regression.fit(X_train, y_train)
# main aim is to find the best fit line
#X_train is independent 
#y_train is dependent feature
```

```python
## print the 13 coefficients and the 1 intercept
print(regression.coef_)
```

```python
print(regression.intercept_) #Here is means, if allthe features are zero then the price of the house is 22.97
```

```python
## Prediction for the test data

reg_pred = regression.predict(X_test)
reg_pred
```

 Assumptions of Linear Regression

```python
plt.scatter(y_test,reg_pred) # Linear model hence model is good
plt.xlabel("Test Truth data")
plt.ylabel("Test predicted data")
```

```python
#residuals
residuals= y_test-reg_pred
```

```python
residuals
```

```python
sns.displot(residuals, kind= "kde") # we get almost gaussian /normal distribution, with little bit of skewed becoz of outliers
```

```python
## scatter plot with predictions and residual
## uniform distribution

plt.scatter(reg_pred,residuals) # uniformly distribution, with no patterns

#term as Homoscedasicity
```

```python
#Performance matrics- 

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test, reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
# compute R square and adjusted R square(adjuscted R square will be less than R square)

from sklearn.metrics import r2_score
score= r2_score(y_test, reg_pred)
print(score)
```

```python
##Adjusted R square
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
